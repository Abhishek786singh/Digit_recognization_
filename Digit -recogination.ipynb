## divide the dataset in feature and label

```python
x = data.iloc[:,1:].values
y = data.iloc[:,0].values
```

## split the dataset into train and test dataset

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

## now use model for fiting the dataset

```python
## i am using Knn
from sklearn.neighbors import KNeighborsClassifier
```

```python
classifier=KNeighborsClassifier(n_neighbors=5,p=2)
classifier.fit(x_train,y_train)
```

## predict the value

```python
y_predict=classifier.predict(x_test)
```

## finding the accuracy

```python

from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,y_predict))
```

```python
y_final=classifier.predict(test)
solution = pd.DataFrame({"ImageId":test.index+1,"Label":y_final})
solution.to_csv("sample_submission.csv",index=False)
```
